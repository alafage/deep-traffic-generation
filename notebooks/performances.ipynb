```python
import pandas as pd

from deep_traffic_generation.core.utils import cumul_dist_plot
```

```python
df = pd.concat(
    [
        pd.read_csv("./outputs/distances/dist_vae_dp.csv").assign(
            generation="VAE", reconstruction="Douglas-Peucker"
        ),
        pd.read_csv("./outputs/distances/dist_vae_nav.csv").assign(
            generation="VAE", reconstruction="Navigational Points"
        ),
        pd.read_csv("./outputs/distances/dist_ref_dp.csv").assign(
            generation="Reference", reconstruction="Douglas-Peucker"
        ),
        pd.read_csv("./outputs/distances/dist_ref_nav.csv").assign(
            generation="Reference", reconstruction="Navigational Points"
        ),
        pd.read_csv("./outputs/distances/dist_gmm_dp.csv").assign(
            generation="GMM", reconstruction="Douglas-Peucker"
        ),
        pd.read_csv("./outputs/distances/dist_gmm_nav.csv").assign(
            generation="GMM", reconstruction="Navigational Points"
        ),
    ]
)
cumul_dist_plot(df, scales=dict(dtw=(0, 1.2e6), sspd=(0, 1.5e4)), domain=["Reference", "VAE", "GMM"])
```
